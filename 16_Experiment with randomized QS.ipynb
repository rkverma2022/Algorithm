```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

RQS_UD = pd.read_csv("07_UD_Complexity_QuickSort.data")
RQS_ND = pd.read_csv("15_ND_Complexity_QuickSort.data")
```

```python
# print(':::::::::::UD_QUICK SORT_ FOR RANDOMISED NUM OF DATA:::::::::::')
# print(RQS_UD)
# print(":::::::::::ND_QS:::::::::::::")
# print(RQS_ND)
RQS_UD = RQS_UD.to_numpy()
RQS_ND = RQS_ND.to_numpy()
```

```python
RQS_UD_Iter = RQS_UD[:, 1] + RQS_UD[:, 2]
RQS_ND_Iter = RQS_ND[:, 1] + RQS_ND[:, 2]
data_UD = RQS_UD[:, 1]
data_ND = RQS_ND[:, 1]

print(data_UD)
fig, ax = plt.subplots(figsize =(5, 3))
ax.hist(data_UD,  color= 'y')
ax.hist(data_ND, color="r")
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(5, 3))
ax.hist(RQS_UD_Iter ,  color= 'y')
ax.hist(RQS_ND_Iter, color="r")
plt.show()
```

# time complexity of quick sort IN AVERAGE CASE
O(n) = k n log(n)                  ; k = O(n)/n log(n)

```python
n_log_n_UD = RQS_UD[:, 0]*(np.log(RQS_UD[:, 0]))
K_UD = RQS_UD[:, 1] / n_log_n_UD

n_log_n_ND = RQS_ND[:, 0]*(np.log(RQS_ND[:, 0]))
K_ND = RQS_ND[:, 1] / n_log_n_ND
plt.figure()


plt.plot(K_UD, marker ='o', color = 'green')
plt.legend(K_UD)
plt.plot(K_ND, marker = '*', color = 'red')
plt.legend(K_ND)
plt.title("Variation in K for UD and ND")
plt.show()
```
