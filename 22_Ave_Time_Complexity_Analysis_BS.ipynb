```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Com_UD = pd.read_csv("21_BS_complexity_18_UD_Normalized.data")
Com_ND = pd.read_csv("21_BS_complexity_19_ND_Normalized.data")
```

```python
Com_UD = Com_UD.to_numpy()
Com_ND = Com_ND.to_numpy()
```

```python
fig, ax = plt.subplots(figsize =(7, 2))
ax.hist(Com_UD)
ax.hist(Com_ND)
plt.show()
```

```python
# number of bucket = b = 10
b = 10
K_UD = (Com_UD[:, 1])/ (Com_UD[:, 0] + b)
K_ND = (Com_ND[:, 1])/(Com_ND[:, 0] + b)
```

```python
plt.figure()


plt.plot(K_UD, marker ='o', color = 'blue')
plt.legend(K_UD)
plt.plot(K_ND, marker = '*', color = 'red')
plt.legend(K_ND)
plt.title("Variation in K for UD and ND")
plt.show()
```
