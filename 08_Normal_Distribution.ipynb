```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_datapoint = int(input("Number of Data Points: \n"))
```

```python
mu, sigma = 0, 0.1 #mean and standard deviation
s = np.random.normal(mu, sigma, number_datapoint)
```

```python
count, bins, ignored = plt.hist(s, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
s = list(s)
```

```python
with open("09_ND_data.data", 'w') as f:
    for i in range(number_datapoint):
        f.write(str(s[i]) +"\n" )
```
